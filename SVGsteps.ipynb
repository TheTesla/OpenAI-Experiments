```python
from gptapilib import ChatBot, printmd
from IPython.display import SVG, display


gen = ChatBot("You are a very creative vector graphics artist! "
              +"Many details! Output valid SVG code only! "
              +"Don't output codeblock formatation!")
chk = ChatBot("Du überprüfst den SVG Code under versuchst leitest "
              +"den Gesprächsparter an eine bessere SVG Grafik zu "
              +"erzeugen!")

a = "Frage deinen Gesprächspartner nach einem sehr hochwertigen " \
    +"SVG, das eine Ente und einen Schwan auf einem See zeigen. " \
    +"Verbessere ihnen Schritt für Schritt."

for i in range(3):
    q = chk.ask(a)
    printmd(f"# CHK \n {q}")
    a = gen.ask(q, temperature=0.7)
    display(SVG(a))
```
